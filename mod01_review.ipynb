### Find the mean, median, and mode of the feature "Age."

```python
Age_mean = Credit['Age'].mean()
print(Age_mean)
```

### Create a subset of Credit that only contains the columns Age, Region, and Income. Call it "Subset."

```python
Subset = Credit[['Age', 'Region', 'Income']]
print(Subset)
```

### From this subset, create two partitions: 
- One that contains records for Ages below 50, and one that contains Ages 50+
- Call them "Younger" and "Older"

```python
Younger = Credit[Credit['Age'] < 50]
print(Younger)
```

```python
Older = Credit[Credit['Age'] >= 50]
print(Older)
```

### Use these partitions to compare the mean and median Income for each age group (less than 50 vs. 50+)

```python
Age_mean = Younger['Age'].mean()
print(Age_mean)
```

```python
Age_mean = Older['Age'].mean()
print(Age_mean)
```

```python
Age_median = Younger['Age'].median()
print(Age_median)
```

```python
Age_median = Older['Age'].median()
print(Age_median)
```

### What does the difference between mean and median suggest about the distribution of income?

Type your answer here.

The difference between mean and median suggests that the distribution of income is non symmetrical and is rather skewed. Looking at the data from both subsets that we have recieved, we can see that the median is greater than the mean indicating that the data is skewed to the left. In other words, this means that there are values of income in both sets of Age that are pulling down the average income.
